```python
import pandas as pd

## heartRate : Frequencia cardiaca, cuantas veces late el corazon por minutos
## QuantumPatternFeature 
##HeartDisease : Enfermedad Cardiaca
## gender 1 = masculino , 0 = Femenino

df = pd.read_csv('HeartPrediction.csv')
```

```python
df.head(10)
```

```python
## Imprimiremos el total de personas Masculinas y Feminas 
sumaMasculino = df['Gender'].value_counts()[1]
sumaFemenino = df['Gender'].value_counts()[0]

print(f" Total Masculino : {sumaMasculino}")
print(f" Total Femenino : {sumaFemenino}")
```

```python
## Imprimiremos la edad media en la que pasa el ataque de corazon
import matplotlib.pyplot as plt

contarEdad = df['Age'].value_counts().sort_index()

plt.figure(figsize=(20, 10))
plt.bar(contarEdad.index, contarEdad.values)
plt.xlabel('Edades')
plt.ylabel('Frecuencia')
plt.title('Distribuci√≥n de Edades en el Dataset')
plt.show()
```

```python
## comentario
 ## "Como podemos observar, la edad donde se realiza mas ataques al conrazon son de 61 a√±os, con una frequencia que pasa de los 17.5"
 ## "Por otra parte, podemos ver que la edad en la que menos riego hay es a los 39 a√±os, por otra parte el rango de edades que padece ataque al corazon es de 50 a 73 a√±os"
```

```python
## A continuacion, voy a ver una de las causas del atauqe al corazon a partir de la maxima presion arterial y a que edad es mas comun tener la presion arterial alta.

maximaPresionArterial = df['BloodPressure'].max()
edadMaximaPresionArterial = df[df['BloodPressure'] == maximaPresionArterial]['Age'].values[0]

print(f"La presi√≥n arterial m√°s alta es: {maximaPresionArterial}")
print(f"La edad en la que se registra esta presi√≥n arterial es: {edadMaximaPresionArterial}")

```

```python
## A partir de los 41 a√±os, podemos observar que la presion arterial es alta.
##¬øEsto por que es debido ? a partir de los 40 tiende a aumentar gradualmente, debido a que las arterias comienzan a perder elasticidad
## y el corazon puede trabajar mas para bombear sangre
```

```python
promedio_presion_por_edad_sexo = df.groupby(['Age', 'Gender'])['BloodPressure'].mean().reset_index()

# Crear subplots
fig, ax = plt.subplots(figsize=(10, 8))

# Crear un diagrama de pastel para cada edad
for age in promedio_presion_por_edad_sexo['Age'].unique():
    data = promedio_presion_por_edad_sexo[promedio_presion_por_edad_sexo['Age'] == age]
    labels = data['Gender']
    sizes = data['BloodPressure']
    
    ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, wedgeprops={'edgecolor': 'black'})
    ax.axis('equal')  # Assegurar que el diagrama de pastel es un c√≠rculo.
    plt.title(f'Presi√≥n Arterial Promedio por Sexo a la Edad de {age}')
    plt.show()
    
```
