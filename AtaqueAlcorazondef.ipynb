```python
import pandas as pd

## heartRate : Frequencia cardiaca, cuantas veces late el corazon por minutos
## QuantumPatternFeature 
##HeartDisease : Enfermedad Cardiaca
## gender 1 = masculino , 0 = Femenino

df = pd.read_csv('HeartPrediction.csv')
```

```python
df.head(10)
```

```python
## Imprimiremos el total de personas Masculinas y Feminas 
sumaMasculino = df['Gender'].value_counts()[1]
sumaFemenino = df['Gender'].value_counts()[0]

print(f" Total Masculino : {sumaMasculino}")
print(f" Total Femenino : {sumaFemenino}")
```

```python
## Imprimiremos la edad media en la que pasa el ataque de corazon
import matplotlib.pyplot as plt

contarEdad = df['Age'].value_counts().sort_index()

plt.figure(figsize=(20, 10))
plt.bar(contarEdad.index, contarEdad.values)
plt.xlabel('Edades')
plt.ylabel('Frecuencia')
plt.title('Distribución de Edades en el Dataset')
plt.show()
```

```python
## comentario
 ## "Como podemos observar, la edad donde se realiza mas ataques al conrazon son de 61 años, con una frequencia que pasa de los 17.5"
 ## "Por otra parte, podemos ver que la edad en la que menos riego hay es a los 39 años, por otra parte el rango de edades que padece ataque al corazon es de 50 a 73 años"
```

```python
## A continuacion, voy a ver una de las causas del atauqe al corazon a partir de la maxima presion arterial y a que edad es mas comun tener la presion arterial alta.

maximaPresionArterial = df['BloodPressure'].max()
edadMaximaPresionArterial = df[df['BloodPressure'] == maximaPresionArterial]['Age'].values[0]

print(f"La presión arterial más alta es: {maximaPresionArterial}")
print(f"La edad en la que se registra esta presión arterial es: {edadMaximaPresionArterial}")

```

```python
## A partir de los 41 años, podemos observar que la presion arterial es alta.
##¿Esto por que es debido ? a partir de los 40 tiende a aumentar gradualmente, debido a que las arterias comienzan a perder elasticidad
## y el corazon puede trabajar mas para bombear sangre
```

```python
import matplotlib.pyplot as plt
import pandas as pd


conteo_por_sexo = df['Gender'].value_counts()

# Calcular los porcentajes
total = sum(conteo_por_sexo)
porcentajes = [100 * size / total for size in conteo_por_sexo]

# Crear el gráfico de pastel
plt.figure(figsize=(6, 6))
plt.pie(porcentajes, labels=conteo_por_sexo.index, autopct='%1.1f%%', startangle=140, wedgeprops={'edgecolor': 'black'})
plt.axis('equal')  # Asegurar que el diagrama de pastel es un círculo
plt.title('Porcentaje de Hombres y Mujeres en el Conjunto de Datos')
plt.show()
    
```
