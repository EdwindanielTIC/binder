```python
#imprimiu quina és la temperatura màxima

maximaTemp = df['Temp'].max()

print(f"la temperatura maxima es {maximaTemp}")
```

```python
##mostreu el mes i el dia (només) que tenim la temperatura màxima

tem_maxima = df['Temp'].max()

fila_maxima = df[df['Temp']==tem_maxima]

mes = fila_maxima['Month'].max()
dia = fila_maxima['Day'].max()

print(f"El mes que tiene la temperatura mas alta es {mes}")

print(f"El dia que tiene la temperatura mas alta es {dia}")
```

```python
##feu un gràfic de punts (scatterplot) per saber si hi ha una relació entre el vent i el Solar.R

import matplotlib.pyplot as plt

plt.scatter(df['Wind'], df['Solar.R'], color='blue', alpha=0.5)

plt.title('Relacion entre el viento y la radiacion Solar')
plt.xlabel('Vent {Wind}')
plt.ylabel('Radiació Solar (Solar.R)')

plt.show()
```

```python
## interpreteu aquest gràfic

# podemos observar que la densidad de puntos estan concentrado en medio, esto signfica que
# tienen rangos comunes entre el viento y la radiacion solar.

# podemos observar tambien que tienen una relacion positiva, porqie forman una linea aascendente de abajo a izquierda
# eso quiere decir que a medida que aumenta el viento, tambien aumenta la radiacion solar.
```

```python
## feu un gràfic de barres que mostri la temperatura mitjana per mes.

# Calcular la temperatura promedio por mes
promedio_mes = df.groupby('Month')['Temp'].mean()

# Crear el gráfico de barras
plt.bar(promedio_mes.index, promedio_mes.values, color='orange')

plt.title('Temperatura promedio por mes')
plt.xlabel('Mes')
plt.ylabel('Temperatura promedio')

# Mostrar el gráfico
```

```python
## interpreteu aquest gràfic 

## cComo podemos obsevar, el mes de julio y agosto es cuando la temperatura esta mas elevada, pasado de los 80
# la minima podemos ver que es mayo con 65 
```
